# Section B – Filtering & Conditional Selection

```python
# 1. Select all students who completed the test preparation course.
# Select rows where test _preparation_course is 'completed'

test_preparation_course = df[df['test preparation course'] == 'completed']
display(test_preparation_course)
```

```python
# 2. Find all students who scored above 70 in math.
# Select rows where test _preparation_course is 'completed'

math_score = df[df['math score'] >70]
display(math_score)
```

```python
# 3. Retrieve female students who scored atleast 65 in all three subjects (use the & operator).

female_above = df[(df['gender'] == 'female') & (df[['math score', 'reading score', 'writing score']] >= 65).all(axis=1)]
display(female_above)
```

```python
# 4. Count how many students have free/reduced lunch.

count_free_lunch = (df['lunch'].value_counts())
display(count_free_lunch)
```

# Section C – Group By & Aggregation

```python
# 1. Compute average math score by gender.

 # Calculate math score by gender
    
math_score = df.groupby('gender')['math score'].mean()
display(math_score)
```

```python
# 2. Compute average scores (math, reading, writing) by parental level of education.
average_score = df.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean()
display(average_score)
```

```python
# 3. Compare averages scores of students with vs without test preparation 
# (note:none refers to students who did not complete the test preparation course).

compare_average_score = df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()
display(compare_average_score)
```

# Section D – Visualization

### Instructions:Use Matplotlib .Include titles and axis labels.

```python
# 1. Ploth histograms for:math score, reading score, writing score

plt.hist(df['math score'], bins=15,edgecolor='black')
plt.title('math Student Perfomance')
plt.xlabel('math score')
plt.ylabel('gender')

plt.show()
```

```python
plt.hist(df['writing score'], bins=15,edgecolor='black')
plt.title('writing Student Perfomance')
plt.xlabel('writing score')
plt.ylabel('gender')

plt.show()
```

```python
# 2. Create a bar plot showing average score by gender.

df.groupby('gender')[['math score', 'reading score', 'writing score']].mean().plot(kind = 'bar')
plt.xlabel('gender')
plt.ylabel('avg_score')
plt.title('student performance')
plt.show()
```
